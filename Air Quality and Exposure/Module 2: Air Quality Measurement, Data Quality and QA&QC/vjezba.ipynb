###Limit of detection
Ten spiked filters were prepared and analyzed to determine the peak area standard deviation. The LOD was then determined as three times the standard deviation (Caseiro et al., 2009, 2007). The detection limits of determined analytes were 0.4, 2.3, 0.1, 1.9, 6.0, 1.1, 1.7, 6.4, 0.9, 1.7, and 1.7 ng m⁻³ for xylitol, arabitol, erythritol, sorbitol, mannitol, mannose, glucose, LG, MNS, galactose, and GA, respectively.
 Method accuracy for each analyte was determined by spiking blank filters and real samples, resulting in recoveries ranging from 85 to 104%.
