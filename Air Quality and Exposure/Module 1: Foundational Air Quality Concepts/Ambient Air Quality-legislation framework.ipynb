# 1. INTRODUCTION

**Air pollution** has been recorded since ancient times. **Archaeological evidence** (such as analyses of ice layers) shows air pollution by lead dating back almost 6,000 years, mostly associated with early metallurgical activities. In **ancient writings**, especially Roman ones, there are records of poor air quality. Roman authors, such as Seneca, described the air in the city as “heavy” and “stifling” due to smoke from furnaces, workshops, and waste burning, which was often associated with dense populations. Unpleasant odors were also a daily occurrence in ancient cities, as evidenced by various accounts from Roman writers. The main causes were open sewage, animal excrement in the streets, rotting waste and food remains, as well as certain trades (leather tanning, textile dyeing).




Until the late Middle Ages and the beginning of the early modern period, people did not attempt to limit or prohibit air pollution through legal means, because pollution was generally very spatially limited.

**The first acts establishing rules for prohibition of air pollution** were adopted in some European countries in the 14th and 15th centuries. However, they were all territorially limited (only to a specific city or municipality) and applied to just one or at most a few sources of pollution.

*For example, the historical development of legal air protection in England began in the beginning of the 14th century, when the growing use of coal in London caused severe smoke and smog problems. In 1306, King Edward I prohibited the use of coal in open fireplaces in London, and Parliament introduced strict penalties for violations. Despite these measures, pollution continued to increase.*

During the following centuries, additional local regulations were introduced, including sanitation laws on waste disposal and urban cleanliness. With the development of industry, the use of energy for economic purposes, as well as increased energy consumption in households, air pollution rose significantly in the second half of the 19th century. Modern legal regulation of air pollution expanded significantly during the second half of the 19th century with industrialization.

**Comprehensive air protection laws were gradually adopted** across European countries during **the second half of the 20th century.** Because air pollution crosses national borders, **international agreements** and **European Union regulations** later became essential instruments for air quality protection.

#2. DEFINITIONS AND TERMS

**Air pollution**: The presence of substances in the atmosphere resulting either from human activity or natural processes, present in **sufficient concentration**, for a **sufficient time** and under circumstances such as to interfere with comfort, health or welfare of persons or the environment.

**Air quality standard**: specified attribute of the air intended to prevent or minimize impacts on humans (e.g. safety and health) and/or the environment; frequently defined statistically by setting **a limit to the concentration** of an air pollutant over **a specified averaging time**; can have legal or advisory status in one or more jurisdictions.

**Determining the thresholds of harmful effects for individual pollutants is a complex task, but it is also a dynamic process, as values are continually reassessed and revised in accordance with new scientific knowledge!**


**The World Health Organization (WHO)** regularly integrates scientific evidence on air pollution’s health impacts and systematically publishes air quality guidelines (AQG) for monitoring air quality, listing harmful substances and recommending limit values with regard to the protection of human health. These guidelines serve as a global target for national, regional and city governments to work towards improving their citizen’s health and generally form the basis for the limit values in the regulations of the countries, including members of European Union, although certain differences do exist.

The recommendations are based on systematic literature reviews and subsequent rigorous evaluation methods as well as extensive consultation with experts and end-users of the guidelines from all regions of the world. They provide recommendations for both short-term (e.g., hourly or daily) and long-term (annual) exposure.

#3. THE WHO AIR POLLUTION GUIDELINES

Numerous studies have been conducted to determine the harmful effects of air pollutants on human health, including epidemiological studies, experiments on cells, tissues, and organisms, tests on experimental animals, and studies involving human volunteers under controlled conditions or individuals who were accidentally exposed to higher concentrations of pollutants. Based on these studies, efforts have been made to determine the concentrations, or thresholds, at which adverse effects on human health occur. However, there are many difficulties in establishing such thresholds.

###**Issues in determining the thresholds of harmful effects:**

- tests under controlled conditions usually include exposure to one or only a few harmful substances, but do not imitate real conditions and are not applicable for long-term exposure
- epidemiological studies reflect real conditions, but due to the numerous factors affecting the studied group, it is difficult to single out the effect of a single pollutant
- in addition to inhalation, a person can also be exposed to the same pollutant in different ways, for example through the skin and by consuming food and drink
- it is not possible to measure all pollutants in the air, only characteristic indicators are measured
- for some pollutants there is no threshold below which there are no harmful effects
-  harmful effects can also occur at concentrations lower than those set by legislation if a person is exposed to them for a long period of time, especially in more sensitive individuals (the elderly, children, and people with chronic illnesses)

#**Global update: 2005**

slika 3.png

While the 1987 AQG relied on the limited epidemiological and toxicological evidence available at the time, the 2005 update incorporated nearly two decades of new research. As large-scale epidemiological studies have demonstrated health effects at much lower concentrations of pollutants than previously recognized, especially for particulate matter, the recommended limit values became significantly stricter. The Global update 2005 addressed the same key “classical” pollutants (PM, SO₂, NO₂, O₃, CO, and lead) as the 1987 AQG, but the 2005 guidelines placed particular emphasis on particulate matter (PM₂.₅ and PM₁₀), recognizing it as one of the most harmful components of air pollution, clearly distinguishing between particle sizes and provided separate guideline values for PM₂.₅ and PM₁₀.


#**The last update: 2021**

The document provide health-based recommended concentration limits for six key air pollutants: particulate matter (PM₂.₅ and PM₁₀), ozone (O₃), nitrogen dioxide (NO₂), sulfur dioxide (SO₂), and carbon monoxide (CO)

#**4. THE AIR QUALITY LEGISLATION IN EU**

By the 1990s, national practices initiated in the 1970s were still being implemented, focusing primarily on the harmonization of technical standards in the interest of the common market. At that time, there was no comprehensive approach to air protection within Europe, but only measures addressing pollution from specific industrial sectors.

The strategy for improving air quality advanced significantly during the 1990s. A major step was the **1996 Framework Directive **(Council Directive 96/62/EC of 27 September 1996 on ambient air quality assessment and management), which introduced a list of 13 atmospheric pollutants to be taken into consideration in the assessment and management of ambient air quality, requirements for monitoring, and action plans in case of exceedances. The 1996 Framework Directive was later supported by “daughter directives” with detailed rules.

In 2001, the EU launched the **CAFE** (Cleaner Air for Europe) **program** to create a long-term, integrated strategy addressing air pollution and its impacts on health and the environment. This led to the 2005 Thematic Strategy and ultimately the 2008 Air Quality Directive, which consolidated earlier legislation and introduced new standards.

Since 2024, the EU ambient air quality legislation framework included the following documents:

**DIRECTIVE 2008/50/EC** OF THE EUROPEAN PARLIAMENT AND OF THE COUNCIL of 21 May 2008 on ambient air quality and cleaner air for Europe

**DIRECTIVE 2004/107/EC** OF THE EUROPEAN PARLIAMENT AND OF THE COUNCIL of 15 December 2004 relating to arsenic, cadmium, mercury, nickel and polycyclic aromatic hydrocarbons in ambient air

**DIRECTIVE 2015/1480/EC** OF THE EUROPEAN PARLIAMENT AND OF THE COUNCIL of 28 August 2015 amending several annexes to Directives 2004/107/EC and 2008/50/EC of the European Parliament and of the Council laying down the rules concerning reference methods, data validation and location of sampling points for the assessment of ambient air quality


These Directives implemented limit/target values recommended by WHO AQG from 2005, but not in full.

Reason:

- values set in the regulations are a compromise between two types of costs, those for preventive air protection and pollution control, and those for medical treatments and removal of harmful effects

- for some pollutants, there is no threshold below which there is no risk to human health, so the WHO has not given guidelines (but only an assessment of the risk at certain concentrations, i.e. a calculation of the probability of cancer in the population for a given exposure)


#**CURRENT LEGAL FRAMEWORK FOR AMBIENT AIR PROTECTION IN THE EU**

**The European Green Deal, December 2019:** The European Commission has committed to further improving air quality and aligning EU air quality standards more closely with the guidelines of the WHO.
**The EU Action Plan:** Zero Pollution Action Plan “Towards zero pollution for air, water and soil” sets out a vision to reduce air, water, and soil pollution by 2050 to levels no longer considered harmful to human health and ecosystems.
**Targets for 2030:**
Two targets are specifically related to air pollution:
•	Improving air quality to reduce the number of premature deaths caused by air pollution by 55%
•	Reducing by 25% the number of ecosystems in the EU where air pollution threatens biodiversity


**DIRECTIVE (EU) 2024/2881** of the European Parliament and of the Council of 23 October 2024 on ambient air quality and cleaner air for Europe (recast)

-	Revision of the previous ambient air quality legislation

-	Improvement the existing legislative framework (e.g., by defining responsibilities and penalties, as well as obligations to inform the public)

-	Provision of support to local authorities in achieving cleaner air through strengthened monitoring, modelling, and planning

Ordinance on Recommended and Limit Air Quality Values (OG 101/96)


The Republic of Croatia joined the European Union on 1 July 2013 as its 28th member state
Croatian legislation and regulations have to be harmonised with EU legislation
EU directives are implemented into the following Croatian documents:
Air Protection Act (OG No. 127/19, 57/22, 136/24)
Regulation on Levels of Pollutants in Ambient Air (OG No. 72/20)
Ordinance on Monitoring Air Quality (OG No. 70/20)
Ordinance on reciprocal exchange of information and reporting on ambient air quality and commitments for implementation of Commission Decision 2011/850/EU (OG No. 3/16)


**The new Air Protection Act will be adopted in 2026 in order to align with the requirements of Directive 2024/2881/EC!**

Limit and target values for SO<sub>2</sub>, NO<sub>2</sub>, O<sub>3</sub>, CO, benzene, PM<sub>10</sub>, PM<sub>2.5</sub> and metals Pb, Cd, Ni, As in PM<sub>10</sub> as well as benzo(a)pyrene (BaP) in PM<sub>10</sub> are the same as in the EU Directive.

Additionally:

-	in Croatia, there ia a limit value for total gaseous mercury (1 µg/m<sup>3</sup> for annual average); EU Directive set an obligation of measurements for Hg as well as the reference measuring method, but not the limit or target value

-	the limit values for ammonia (NH<sub>3</sub>) , hydrogen sulphide (H<sub>2</sub>S), and mercaptans (R-SH) remained unchanged from previous Croatian legislation; limit values were set considering quality of life (based on the odour detection threshold; not on the basis of the appearance of harmful health effects)

-	the issue of offensive odours is currently not covered by any other Croatian regulation and is not harmonized within EU


###**The case of Zagreb, Croatia**

How will the new limit values reflect on air quality assessment? in the case of Zagreb, Croatia?

What if we take into account the values recommended by the WHO?




Data from 6 monitoring stations of the local air quality monitoring network funded by the City of Zagreb were analyzed:

-	period: 2018 – 2022

-	pollutants: NO<sub>2</sub>, PM<sub>10</sub>, PM<sub>2.5</sub>


Additionally, PM<sub>10</sub> and PM<sub>2.5</sub> gravimetric data from selected monitoring stations of the Croatian network for continuous air quality monitoring for the same period.

For the period 2018-2022, air quality in Zagreb with regard to SO<sub>2</sub>, CO, benzene and metals (Pb, Cd, As, Ni) in PM<sub>10</sub> was found to be satisfactory and in accordance with all of the aforementioned standards.

slika 5_1.png

Taking into account current legislation, levels of PM<sub>10</sub> exceeded limit values only at some monitoring stations.


Taking into account limit values that will apply after 1 January 2030, the air would be considered polluted at some Zagreb stations regarding NO<sub>2</sub>, and at all stations regarding PM<sub>10</sub>.

Considering PM<sub>2.5</sub> the air quality would be considered satisfactory only at monitoring stations Rijeka-2 and Plitvička jezera.

Considering WHO guideline values, the air quality **was unsatisfactory at all stations for all analyzed pollutants (NO<sub>2</sub>, PM<sub>10</sub>, and PM<sub>2.5</sub>)**




Although decreasing trends of pollutant concentrations were observed at most of the stations, it seems difficult to achieve the new EU objectives for PM10 and PM2.5 before 2030.
