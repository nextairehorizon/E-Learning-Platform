## 2. Pandas DataFrames

In this notebook, we move from arrays to tables. **Pandas DataFrames** are one of the most important tools for scientific data manipulation in Python.

A DataFrame is similar to a spreadsheet: it has rows, columns, and labels. It is especially useful when measurements come with metadata such as station name, timestamp, or sensor type.

## 2.1 Importing Pandas and Building a Table

Pandas is commonly imported with the alias `pd`. We will also import NumPy because it is often used together with Pandas.


```python
import pandas as pd
import numpy as np
```

```python
data = {
    "station": ["A", "A", "B", "B", "C", "C"],
    "date": ["2026-05-01", "2026-05-02", "2026-05-01", "2026-05-02", "2026-05-01", "2026-05-02"],
    "temperature": [18.4, 19.1, 17.8, 18.3, 20.2, 20.5],
    "humidity": [62, 60, 68, 66, 55, 54],
    "rain_mm": [0.0, 1.2, 3.4, 0.8, 0.0, 0.0]
}

df = pd.DataFrame(data)
df
```

Each column contains one variable, and each row represents one observation.

This is a very common structure in research data: one row per sample, one column per feature or measured quantity.

## 2.4 Selecting Columns and Rows

We can select:

- one column,
- several columns,
- rows by position,
- rows by label or condition.


```python
df["temperature"]
```

```python
df[["station", "temperature", "humidity"]]
```

```python
df.iloc[0:3]
```

```python
df.loc[0:2, ["station", "rain_mm"]]
```

- `iloc` selects by integer position.
- `loc` selects by labels and column names.

For beginners, this distinction is worth learning early because it appears frequently in Pandas code.


## 2.5 Filtering Rows

Filtering keeps only rows that satisfy a condition.


```python
warm_days = df[df["temperature"] > 18.5]
warm_days
```

```python
dry_days = df[df["rain_mm"] == 0]
dry_days
```

```python
warm_and_dry = df[(df["temperature"] > 18.5) & (df["rain_mm"] == 0)]
warm_and_dry
```

This type of filtering is common in environmental data analysis. For example, a researcher may want to inspect only dry days, only high-humidity periods, or only observations from one station.


## 2.6 Sorting Data

Sorting can help us identify the largest or smallest values quickly.


```python
df.sort_values(by="temperature", ascending=False)
```

```python
df.sort_values(by=["station", "date"])
```

Sorting by multiple columns is useful when we want a logical order, for example station first and date second.


## 2.7 Creating New Columns

A DataFrame becomes more useful when we derive new variables from existing ones.


```python
df["temperature_f"] = df["temperature"] * 9/5 + 32
df["dry_day"] = df["rain_mm"] == 0
df["humidity_fraction"] = df["humidity"] / 100

df
```

These transformations are vectorized, which means they are applied to the whole column at once.


## 2.8 Renaming and Reordering Columns

Researchers often rename columns to make a dataset easier to read or more consistent with later analysis.


```python
renamed = df.rename(columns={"rain_mm": "rainfall_mm"})
renamed = renamed[["station", "date", "temperature", "temperature_f", "humidity", "humidity_fraction", "rainfall_mm", "dry_day"]]
renamed.head()
```

This step is not always necessary, but it is useful when combining data from different sources that use slightly different column names.


## 2.9 Grouping and Aggregation

A common task is to compute summaries for groups of rows. In environmental data, we may want average measurements per station.


```python
df.groupby("station")["temperature"].mean()
```

```python
df.groupby("station")[["temperature", "humidity", "rain_mm"]].mean()
```

```python
summary = df.groupby("station").agg({
    "temperature": ["mean", "max"],
    "humidity": "mean",
    "rain_mm": "sum"
})

summary
```

Grouping and aggregation are powerful because they reduce many rows into a smaller summary table.
