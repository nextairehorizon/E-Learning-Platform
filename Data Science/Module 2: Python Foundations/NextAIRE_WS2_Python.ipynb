⚠️ **Gotcha — copying.** Assigning a list does **not** copy it.

````python
a = [1, 2, 3]
b = a          # b is the SAME list
b.append(99)
b        # surprise! a changed too
````

````python
a
````

````python
a = [1, 2, 3]
a
````

````python
b = a.copy()   # independent copy
b.append(99)

````

````python
a, b
````

## 6 · Functions

A function packages a piece of logic so we can reuse it.

```python
def name(arguments):
    """docstring — what it does"""
    ...
    return result
```


````python
def double(x):
    """Return x multiplied by 2."""
    y = x * 2
    return y
````

````python
double(3.14)
````

Once defined, a function is just another value — we can inspect its type and its help.

````python
type(double)
````

````python
help(double)
````

### A real example — pH of a solution

````python
import math
````

````python
-math.log10(0.0001)
````

````python
def ph(c):

    phval = -math.log10(c)
    return phval
````

````python
help(ph)
````

````python
ph(0.001)
````

### Multiple arguments + default values

````python
def bmi(weight_kg, height_m):
    return weight_kg / height_m ** 2

bmi(84, 1.80)
````

Defaults let us call a function with fewer arguments.

> Dilution: c₁·V₁ = c₂·V₂


````python
def diluter(c1, V1=1, V2=1):
    """Return concentration after dilution."""
    return c1 * V1 / V2

diluter(2)          # uses defaults
````

````python
diluter(2, V1=1, V2=4)   # keyword arguments — order-independent
````

**Gotcha** — after a keyword argument, you can't go back to positional.

```python
diluter(c1=2, 1, 4)   # SyntaxError
```
