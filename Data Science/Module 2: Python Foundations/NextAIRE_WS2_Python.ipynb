### `break` and `continue`

`break` exits the loop, `continue` skips to the next iteration.

```python
readings
```

```python
for pm_value in readings:
    if pm_value > 50:
        break
    print(pm_value)
```

⚠️ `None` means a missing value (e.g. broken sensor). Test for it with `is None` — **not** `== None`.

```python
for pm in [12, None, 28]:
    print(pm, ' 1st print')
    if pm is None:
        continue
    print(pm, ' 2nd print')
    print('======')
```

## 4 · `while` loops

`while` keeps running **as long as** a condition is `True`. Useful when we don't know in advance how many iterations we need.

```python
pm = 80
while pm > 25:
    pm = pm - 10
    print(pm)
```

⚠️ **Gotcha** — forget to update the variable and the loop runs forever. If that happens, hit the ⏹ stop button in Jupyter.

Simulated air-filter run: keep halving the concentration until we reach safe air.

```python
pm = 1000
steps = 0
while pm > 25:
    pm = pm / 2
    steps += 1
    print('pm: ', pm, ' steps: ', steps)
```

```python
pm, steps
```
